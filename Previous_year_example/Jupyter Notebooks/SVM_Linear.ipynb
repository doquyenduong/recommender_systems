### SVM with Sigmoid Kernel
Source of Code: https://stackabuse.com/implementing-svm-and-kernel-svm-with-pythons-scikit-learn/

```python
import pandas as pd
import sklearn
```

Here we import the encoded dataset

```python
df = pd.read_csv('Deezer_encoded_Big.csv')
df
```

We drop an unnnecessary column and take a sample of 1% of the dataset.

```python
df= df.drop(columns=['Unnamed: 0'])
df
```

```python
df1 = df.sample(frac =0.01) 
df1
```

Here we split the dataset into features and label

```python
features2 = df1.iloc[:, :-1].values
label2 = df1.iloc[:, 13].values
```

Here we split the dataset into train and validation data. We do a split of 80/20.

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(features2, label2, test_size=0.20, random_state=109)
```

We define a SVM classifier and define the kernel as linear.

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

Finally, we plot the accuracy of that model. We can observe that we reached an accuracy of 69%

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred)
```
